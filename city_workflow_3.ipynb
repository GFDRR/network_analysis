```python
def calculate_shortest_paths(origins_gdf, POIs, G, target_crs = 'epsg:32646'):
    
    origins_gdf = gpd.GeoDataFrame(origins_gdf.copy())
    
    origins_gdf = gn.pandana_snap(G, origins_gdf, add_dist_to_node_col=True,
                              source_crs='epsg:4326', target_crs = target_crs)
    
    POIs = gn.pandana_snap(G, POIs, add_dist_to_node_col=True,
                               source_crs= 'epsg:4326', target_crs = target_crs)
    
    OD = gn.calculate_OD(G, list(origins_gdf.NN), list(POIs.NN), 0, weight='time')
    OD_df = pd.DataFrame(OD, index = origins_gdf.pointid, columns = POIs.index)
    
    origins_gdf['travel_time_road'] = OD_df.min(axis = 1)
    
    origins_gdf['destination_ID'] = OD_df.idxmin(axis=1)
    destination_walk_distances_dict = dict(zip(set(POIs.index),set(POIs.NN_dist)))
    origins_gdf['destination_walk_distance'] = origins_gdf.destination_ID.map(destination_walk_distances_dict)
    
    origins_gdf['walk_time_origin'] = ((origins_gdf.NN_dist / 1000 ) / 4.5) * 60 * 60      # in seconds
    origins_gdf['walk_time_destination'] = ((origins_gdf.destination_walk_distance / 1000 ) / 4.5) * 60 * 60
    
    origins_gdf['travel_time'] = (origins_gdf['travel_time_road'] +
                                  origins_gdf['walk_time_origin']) / 60   # In minutes
    origins_gdf.dropna(inplace=True)
    return(origins_gdf)
```
